## Climbing Stairs

* https://leetcode.com/problems/climbing-stairs/
***
* Time Complexity:
    - naive solution: O(2$^{n}$)
        * for every step, you're looking at climbingStairs(step - 1) and climbStairs(step - 2) to calculate the answer
        * the solution doesn't keep track of things we already calculated so even for easy solutions like climbingStairs(3), we'd still have to calculate it again
        * this leads to 2$^{n}$ nodes in the recursion tree
    - top-down memoized dp: O(n)
        * any previously calculated values are already stored in the memo hash table
        * so you'll only have to do a calculation for each value of n once
        * therefore it is O(n)
    - bottom-up dp: O(n)
        * our for loop here only iterates until n so it is O(n)
* Space Complexity:
    - naive solution: O(n)
        * you don't calculate step - 1 and step - 2 in parallel so any calls to step - 1 are completed first before step - 2
        * at most, you'll have n function calls b/c if you calculated step - 1 for step = 3, you'll have 3 function calls to make and you'll never exceed that if you calculated step - 2
    - top-down memoized solution: O(n)
        - the memo table will have at most O(n) items in it and since you're still using recursion, you won't have more than O(n) function calls since it'll return quickly if the steps are in the table
    - bottom-up dp: O(1)
        - you don't use recursion and you only keep track of 2 variables

```javascript
/**
 * @param {number} n
 * @return {number}
 */

// naive solution
var climbStairs = function(n) {
    // base case
    // 0 ways to climb 0 steps, 1 way to climb 1 step
    if (n <= 2) return n;
    
    return climbStairs(n - 1) + climbStairs(n - 2);
}

// top-down memoized dp
var climbStairs = function(n) {
    if (n <= 2) return n;
    
    const memo = {
        0: 0,
        1: 1,
        2: 2
    };
    
    const traverse = (steps) => {
        if (steps < 0) return 0;
        
        else if (memo[steps] !== undefined) {
            return memo[steps];
        }
        
        memo[steps] = traverse(steps - 1) + traverse(steps - 2);
        
        return memo[steps];
    }
    return traverse(n);
};

// bottom-up dp
var climbStairs = function(n) {
    if (n <= 2) return n;
    
    let last1 = 2;
    let last2 = 1;
    
    for (let i = 2; i < n; i++) {
        [last1, last2] = [last1 + last2, last1];
    }
    
    return last1;
}
```

## Min Cost Climbing Stairs

* https://leetcode.com/problems/min-cost-climbing-stairs/
***
* Time Complexity: O(n)
    - you traverse through the entire array and you only do work on the 2 variables
* Space Complexity: O(1)
    - only 2 variables are created and used and there's no recursion either
***
* any problems that ask you to look at the next 1 and 2 items in the array should have a very similar structure to fibonacci
* in this case, we move from 2 ... n, and starting from index 2, we look at i - 1 and i - 2
    - as we do this, we look at the minimum of those two and add them to the current value at i
* you don't NEED to look at i + 1 and i + 2 while you loop from 0 ... n
    - just use the base case and try to figure out how you want to move
    - in this case, if we only have 2 numbers, we only need to compare them and take the min
    - if we add a 3rd number to the list, we just take the minimum of the last 2 we had and add it to i, so Math.min(arr[i - 1], arr[i - 2])
    - then when everything is traversed, we take the minimum of the 2 minimums we calculated

```javascript
/**
 * @param {number[]} cost
 * @return {number}
 */
var minCostClimbingStairs = function(cost) {
    if (cost.length === 2) return Math.min(cost[0], cost[1]);
    
    let by1 = cost[1];
    let by2 = cost[0];
    
    for (let i = 2; i < cost.length; i++) {
        let min = Math.min(by1, by2);
        [by1, by2] = [cost[i] + min, by1];
    }
    
    return Math.min(by1, by2);
};
```

## House Robber

* https://leetcode.com/problems/house-robber/
***
* Time Complexity: O(n)
    - you're basically just looping from 2 ... n
* Space Complexity: O(1)
    - only using 2 pointers x and y to accumulate the maxes as we go along
***
* similar to Fibonacci
* start from the base case:
    - [] = 0 b/c no houses
    - [1] = 1 b/c just 1 house to rob
    - [1, 2] = 2 since they're adjacent, we can either rob house 1 or house 2, not both so we take the max of those 2
    - [1, 2, 3] = 4 b/c, starting at index 1, we can either get loot from house[i] + houses we've robbed[i - 2] OR we just rob the previous house
    - so essentially, we rob the previous house OR the previous subarray of houses
    - that's why we need to do [x, y] = [Math.max(y, nums[i] + x]
    - and also the reason why we need to set y = Math.max(nums[0], nums[1]);
        * b/c it is the max of the previous subarray we've seen

```javascript
/**
 * @param {number[]} nums
 * @return {number}
 */
var rob = function(nums) {
    if (nums.length <= 1) {
        return nums.length ? nums[0] : 0;
    }
    
    let x = nums[0];
    let y = Math.max(nums[0], nums[1]);
    
    for (let i = 2; i < nums.length; i++) {
        [x, y] = [y, Math.max(y, nums[i] + x)];
    }
    
    return y;
};
```

## House Robber II

* https://leetcode.com/problems/house-robber-ii/description/
***
* Time Complexity: O(n)
    - essentially just calling House Robber twice and House Robber is O(n)
* Space Complexity: O(1)
    - you only keep a couple of variables in memory
***
* the problem states that the first and last houses are technically adjacent to each other
* thus, we have to call House Robber twice on 2 sets of houses here:
    - if a neighborhood = [0 ... n]
    - then we call House Robber on the first set from [0 ... n - 1]
        * exclude the last house in this set
    - and then the second set from [1 ... n ]
        * exclude the first house in this set
* we then return the max of the 2 sets

```javascript
/**
 * @param {number[]} nums
 * @return {number}
 */

 // bottom-up dp
var rob = function(nums) {
    if (nums.length === 1) return nums[0];
    if (nums.length === 2) return Math.max(nums[0], nums[1]);

    const robbery = (start, end) => {
        const len = end - start + 1;
        if (len === 1) return nums[start];
        if (len === 2) return Math.max(nums[start], nums[end]);

        let x = nums[start];
        let y = Math.max(x, nums[start + 1]);

        for (let i = start + 2; i <= end; i++) {
            [x, y] = [y, Math.max(y, x + nums[i])];
        }

        return Math.max(x, y);
    }

    // you essentially do House Robber twice
    // since the first and the last houses are adjacent in this case
    // we instead call house robber on all houses except the last house
    // then we call house robber on all houses except the first house
    const leftHouses = robbery(0, nums.length - 2);
    const rightHouses = robbery(1, nums.length - 1);
    return Math.max(leftHouses, rightHouses);
};
```

## Longest Palindromic Substring

* https://leetcode.com/problems/longest-palindromic-substring/description/
***
* Time Complexity:
    - naive: O(n$^{3}$)
        * looping through every substring = O(n$^{2}$)
        * checking if each substring is a palindrome = O(n)
    - dp: O(n$^{2}$)
        * create the dp table = O(n$^{2}$)
        * looping through every substring = O(n$^{2}$)
        * checking if each substring is a palindrome = O(1) b/c we do 3 O(1) checks:
            - if s[start] === s[end]
            - if end - start === 1
                * this is for cases where we have even palindromes, e.g. cbbd, bb = even palindrome
            - if dp[start + 1][end - 1] === true
                * if we know that the start and end are equal, then check the middle portion
                * if s[1] === s[5], then check if s[2]...s[4] is a palindrome
    - two pointer: O(n$^{2}$)
        * loops from first to last char in s = O(n)
        * extends left/right pointers outward from the current index until a palindrome is no longer possible = O(n)
* Space Complexity:
    - naive: O(1)
        * only keeps a couple of variables
    - dp: O(n$^{2}$)
        * creates a dp table of size n x n
    - two pointer: O(1)
        * only keeps a couple of variables
***
* dp algorithm:
    - the naive algorithm has us checking the ENTIRETY of the substring if it's a palindrome but we don't actually need to do that
    - we know that if the first + last characters of a substring are equal and if the middle characters are a palindrome, then the entire thing is a palindrome
        * e.g. we have a substring aba. this is a palindrome
        * if we attach 2 Cs around it, then we get cabac
            - the first and last characters are the same. c === c
            - and we already know that aba is a palindrome
            - therefore, cabac is a palindrome
            - these are all O(1) checks
    - so our recurrence relation is now:
        * substring[i,j] = s[i] === s[j] and dp[i + 1][j - 1] === true
        * dp[i + 1][j - 1] represents the substring between i and j, not inclusive of s[i] and s[j]
    - as for how we should loop through this algorithm, we must make a conscious choice about where we place our start + end pointers
        * __WE SHOULD ALWAYS LOOP THE TABLE IN THE OPPOSITE DIRECTION OF OUR RECURRENCE RELATION__
        * e.g. our recurrence relation needs [i + 1] and [j - 1]
            - therefore we need to loop from [i ... 0] and [j ... n]
            - this is so that when we try to find dp[i + 1][j - 1], we will always find a SOLVED answer, meaning we've already solved for that substring and know its answer
    - then when we've found our answer, we return the slice of it from the string
        * we save on time here b/c we only do this operation once at the end rather than doing this everytime we find a longer palindrome
        * we also only need 2 variables for this: maxLen and subStart
            - subStart = start of the palindromic substring
            - and we only need to slice from subStart ... subStart + maxLen since slice goes up to but not including the last index
* two pointer:
    - the previous 2 algorithms found the longest palindrome by moving from the outside in but this algorithm focuses on going from the inside out
    - we know that a single character, e.g. "a", is a palindrome. it can be read backwards and forwards the same
    - thus using this knowledge, we can instead expand ourselves out from that singular point to the left and right
        * <-- a -->
        * if the characters to the left and right of our palindrome match then we can keep going until they don't match anymore
    - so to do this, we essentially treat each character in the substring as a point from which we can expand outwards from
        * so we loop from 0...n
    - for each of these start points, we have to do 2 things:
        1. identify if there are any duplicates next to us
            * this handles the case where we have an even palindrome, e.g. caab, palindrome = aa
        2. then we extend out until s[start] !== s[end]

```javascript
/**
 * @param {string} s
 * @return {string}
 */

 // O(n^3) naive algorithm
const isPalindrome = (start, end, s) => {
    while (start <= end) {
        if (s[start] !== s[end]) {
            return false;
        }
        start++;
        end--;
    }

    return true;
}

var longestPalindrome = function(s) {
    if (s.length === 1) return s;

    let maxLen = 0;
    let str = "";

    for (let i = 0; i < s.length; i++) {
        for (let j = i; j < s.length; j++) {
            
            if (!isPalindrome(i, j, s)) continue;

            const runningLen = j - i + 1;
            if (runningLen > maxLen) {
                maxLen = runningLen;
                str = s.slice(i, j + 1);
            }
        }
    }

    return str;
};

/**
 * dp algorithm
 * Time: O(n^2)
 * Space: O(n^2)
 */
var longestPalindrome = function(s) {
    const n = s.length;

    // create dp table
    const dp = Array.from({length: n}, () => Array.from({length: n}, () => false));

    // fill up the table for single chars in s
    // b/c we know that single letters are palindromes
    for (let i = 0; i < n; i++) {
        dp[i][i] = true;
    }

    // we loop from the bottom right side to the top left
    // reason being, our recurrence relation is this:
    // substring[i, j] = s[i] === s[j] && dp[i + 1][j - 1] === true
    // so we need to know what dp[i + 1]
    // so filling the table up from the bottom is the best place to get it
    // we also go from i + 1 to n for j or else we go out of bounds

    // in essence, we fill in the opposite direction of what the recurrence relation needs
    let maxLen = 1;
    let subStart = 0;

    for (let start = n - 1; start >= 0; start--) {
        for (let end = start + 1; end < n; end++) {
            // check the recurrence relation
            if (s[start] === s[end]) {
                if (end - start === 1 || dp[start + 1][end - 1]) {
                    dp[start][end] = true;
                    const subLen = end - start + 1;
                    if (subLen > maxLen) {
                        maxLen = subLen;
                        subStart = start;
                    }
                }
            }
        }
    }

    // only need where the palindrome starts and the maxLen of the palindrome
    // we perform this operation once when we've found the max palindrome rather than
    // setting it everytime in the loop which would be O(n) multiplied rather than added
    // to the time complexity
    return s.slice(subStart, subStart + maxLen);
}

/**
 * two pointer
 * Time: O(n^2)
 * Space: O(1)
 */
 var longestPalindrome = function(s) {
     const n = s.length;

    let maxLen = 1;
    let subStart = 0;

     for (let i = 0; i < n; i++) {
         let end = i;
         while (end < n && s[i] === s[end]) {
             end++;
         }


         let start = i - 1;
         while (start >= 0 && end < n && s[start] === s[end]) {
             start--;
             end++;
         }

        // reason why this is end - start - 1
        // is b/c our while loop exits when s[start] !== s[end]
        // so if we found the longest substring, we actually don't stop there
        // we keep going
        // so instead of doing end - start + 1, we do end - start - 1
        
        // we originally add 1 b/c the length of the palindrome is 1 less due to
        // the indexing
        // but since we need to account for the length being extra by 2 due to the loop
        // we do end - start - 1. which should get us to the actual length

        // in essence:
        // end - start + 1 = len 
        // end - start = len - 1
        // end - start - 1 = len - 2 = WHAT WE WANT
         const subLen = end - start - 1;
         if (subLen > maxLen) {
             maxLen = subLen;

             // the same logic applies from above
             // since we stop when s[start] !== s[end]
             // if we reach the longest palindrome, we would've done an extra start--
             // so to account for that, we do start + 1
             subStart = start + 1;
         }
     }

     return s.slice(subStart, subStart + maxLen);
 }
```

## Palindromic Substrings

* https://leetcode.com/problems/palindromic-substrings/description/
***
* Time Complexity:
    - dp:  O(n$^{2}$)
        * have to create the dp table of size n x n and then filling that table up
    - two pointer:  O(n$^{2}$)
        * loop from 0...n and for each index, we also expand outwards at most n/2 times
* Space Complexity:
    - dp:  O(n$^{2}$)
        * need to store a dp table of size n x n
    - two pointer: O(1)
        * only needs to keep a couple of variables
***
* dp:
    - similar to longest palindromic substring
    - we create a dp table and fill it up
        * dp[i][i] = always true b/c a single character is a palindrome
        * then loop from bottom-right --> top left and fill up the table
        * our recurrence relation:
            - substring[i,j] = s[i] === s[j] and dp[i + 1][j - 1] === true
    - this time, we don't have to worry about calculating the longest substring
    - we just need to increase the count every time that condition passes
* two pointer:
    - for each character in the string, we can expand 2 pointers outwards as long as those 2 are equal
    - we also take care of duplicates next to each by moving the end pointer to the right until there are no more duplicates
    - everytime we expand the pointers, we increment the count

```javascript
/**
 * @param {string} s
 * @return {number}
 */

 // dp algorithm
var countSubstrings = function(s) {
    const n = s.length;

    // create dp table
    const dp = Array.from({length: n}, () => Array.from({length: n}, () => false));

    // fill up table for single chars b/c they're palindromes
    for (let i = 0; i < n; i++) {
        dp[i][i] = true;
    }

    // there will always be n palindromes b/c a single char = palindrome
    let count = n;
    // loop starting from bottom-right --> top-left
    for (let start = n - 1; start >= 0; start--) {
        for (let end = start + 1; end < n; end++) {
            if (s[start] === s[end]) {
                if ((end - start === 1) || dp[start + 1][end - 1]) {
                    dp[start][end] = true;
                    count++;
                }
            }
        }
    }

    return count;
};

// two pointer
var countSubstrings = function(s) {
    const n = s.length;

    let count = 0;
    // loop through each char
    for (let i = 0; i < n; i++) {
        let end = i;

        // expand the end until all duplicates are accounted for
        while (end < n && s[i] === s[end]) {
            end++;
            count++;
        }

        // start should be a pointer before i
        let start = i - 1;

        while (start >= 0 && end < n && s[start] === s[end]) {
            start--;
            end++;
            count++;
        }
    }

    return count;
}
```

## Decode Ways

* https://leetcode.com/problems/decode-ways/description/
***
* Time Complexity: O(n)
    - only needs to traverse through the entire string once
    - other operations are O(1)
* Space Complexity: O(1)
    - only needed access to dp[i - 1] and dp[i - 2] which can be stored easily in 2 variables
***
* basically fibonacci with some harder conditions to handle
* you just add dp[i - 1] and dp[i = 2] together to get your answer since adding a new character does not actually create a totally new subset
    - it instead adds the character to already existing subsets
    - e.g. 123 = [1,2] + 3, [12] + 3, [1] + 23

```javascript
/**
 * @param {string} s
 * @return {number}
 */

 /**
  * dp algorithm
  * Time: O(n)
  * Space: O(n)
  */
var numDecodings = function(s) {
    if (s[0] === "0") return 0;
    if (s.length === 1) return 1;

    const n = s.length;
    const dp = [1];

    if (Number(`${s[0]}${s[1]}`) <= 26) {
        dp.push(s[1] === "0" ? 1 : 2);
    }
    else {
        dp.push(s[1] === "0" ? 0 : 1);
    }

    for (let i = 2; i < n; i++) {
        dp[i] = 0;

        // if it's anything but a 0, then initialize it to previous
        if (s[i] !== "0") {
            dp[i] += dp[i - 1];
        }

        // if s[i - 1] !== "0"
        // if s[i - 1:i] <= 26
        if (s[i - 1] !== "0" && Number(`${s[i - 1]}${s[i]}`) <= 26) {
            dp[i] += dp[i - 2];
        }
    }
    return dp[n - 1];
};

 /**
  * dp algorithm
  * Time: O(n)
  * Space: O(1)
  */
var numDecodings = function(s) {
    if (s[0] === "0") return 0;
    if (s.length === 1) return 1;

    const n = s.length;
    let x = 1;
    let y;

    if (Number(`${s[0]}${s[1]}`) <= 26) {
        y = s[1] === "0" ? 1 : 2;
    }
    else {
        y = s[1] === "0" ? 0 : 1;
    }

    for (let i = 2; i < n; i++) {
        let temp = y;

        y = 0;
        // if it's anything but a 0, then initialize it to previous
        if (s[i] !== "0") {
            y += temp;
        }

        // if s[i - 1] !== "0"
        // if s[i - 1:i] <= 26
        if (s[i - 1] !== "0" && Number(`${s[i - 1]}${s[i]}`) <= 26) {
            y += x;
        }

        x = temp;
    }
    return y;
};
```

## Coin Change

* https://leetcode.com/problems/coin-change/description/
***
* Time Complexity:
    - naive: O(n$^{m}$), m = amount, n = # of coins
        * each amount represents a node and each node will spawn n nodes
    - top-down memo: O(m x n)
        * for each amount, we check the min. if we subtract each coin value from it
        * the dp array will improve performance because we will only perform operations for an amount once
    - bottom-up dp: O(m x n)
        * for every coin value in the coins array, we also loop from coin...amount
* Space Complexity:
    - naive: O(m), m = amount
        * the recursion will place O(m) functions in stack
        * e.g. if you have amount = 100, and coins = [1,2], then you will go through 100 functions before returning
    - top-down memo: O(m)
        * requires O(m) space for the dp array
        * requires O(m) space for the recursive functions
    - bottom-up dp: O(m)
        * requires O(m) space for the dp array
***
* naive algorithm:
    - base case: if the current amount <= 0, we return 0
        * need 0 coins to make 0
    - we set the amount to ${\infty}$
    - for the current amount, we subtract every coin from that amount and see what the min is
        * so our __recurrence relation is: dp[amount] = min(dp[amount], 1 + dp[amount - coin])__
    - once we have found the min, we +1 to it and return it
        * the +1 represents the extra coin that was needed to get to the current amount
        * i.e. 1 + min(curAmount - coin)
    - we can also choose to sort the coins array and break from the loop early if the current amount - coin is < 0
        * reason being, if we sort it, we know that the coins will only increase in value and we will always satisfy that break condition
* top-down memo algorithm:
    - we actually do repeat work because current amount - coin will have us traverse on an amount we've seen before
    - therefore, we can initialize a dp array
        * dp's length is equal to the amount where each index represents the min. # of coins needed to create it
        * dp[0] = 0 b/c we can only get to 0 with 0 coins
    - then during the recursion, if we have already seen the dp[amount] we can return it
    - otherwise, we can traverse through all of the coins for it, find the min, and set dp[current amount] = min + 1
* bottom-up dp algorithm:
    - we essentially do the same thing but this time we start from 0 and build up
    - we know that it takes 0 coins to make 0 so that's our starting point
    - for each coin value, we loop from coin...amount
        * while we do this, we can figure out how many coins are needed to get to that certain amount
        * as we loop through every coin, the # of coins needed for each amount will eventually get smaller and smaller
        * e.g. amount = 5, [1, 2]
            - coin = 1, dp = [0, 1, 2, 3, 4, 5]
            - coin = 2, dp = [0, 1, 1, 2, 2, 3]

```javascript
/**
 * @param {number[]} coins
 * @param {number} amount
 * @return {number}
 */

// naive recursive algorithm
 var coinChange = function(coins, amount) {
    if (amount === 0) return 0;

    // sort it first to break out of the loop earlier
    coins.sort((a, b) => a - b);

    const traverse = (curAmount) => {
        if (curAmount <= 0) return 0;

        // find the min for each amount first
        let min = Number.POSITIVE_INFINITY;
        for (let coin of coins) {
            if (curAmount - coin < 0) break;
            min = Math.min(min, traverse(curAmount - coin));
        }

        return min + 1;
    }

    const res = traverse(amount);
    return isFinite(res) ? res : -1;
};

// 
 // Time: O(m * n), m = amount, n = # of coins
 // in the case where our only coin = 1 and we have to create a larger number
 // Space: O(m) b/c dp's length is equal to the amount
 // b/c each index of the dp represents the min coins needed to create the amount
var coinChange = function(coins, amount) {
    if (amount === 0) return 0;
    const dp = [0];

    // sort it first to break out of the loop earlier
    coins.sort((a, b) => a - b);

    const traverse = (curAmount) => {
        if (dp[curAmount]) return dp[curAmount];
        if (curAmount <= 0) return 0;

        // find the min for each amount first
        let min = Number.POSITIVE_INFINITY;
        for (let coin of coins) {
            if (curAmount - coin < 0) break;
            min = Math.min(min, traverse(curAmount - coin));
        }

        // the + 1 accounts for the coin needed to reach curAmount
        dp[curAmount] = min + 1;
        return dp[curAmount]
    }


    // if none of the coins can make the amount
    // then it'll show up as Number.POSITIVE_INFINITY in the dp
    const res = traverse(amount);
    return isFinite(res) ? res : -1;
};

// bottom-up dp
// Time: O(m * n), m = amount, n = # of coins
// Space: O(m), m = amount
var coinChange = function(coins, amount) {
    const dp = Array.from({length: amount + 1}, () => Number.POSITIVE_INFINITY);
    dp[0] = 0;

    // loop through all coins
    for (let coin of coins) {
        if (coin > amount) break;
        for (let i = coin; i <= amount; i++) {
            dp[i] = Math.min(dp[i], 1 + dp[i - coin]);
        }
    }

    return isFinite(dp[amount]) ? dp[amount] : -1;
}
```

## Maximum Product Subarray

* https://leetcode.com/problems/maximum-product-subarray/description/
***
* Time Complexity:
    - naive: O(n$^{2}$)
    - dp: O(n)
* Space Complexity:
    - naive: O(1)
    - dp: O(1)
***
* naive:
    - look at all subarrays, calculate their products along the way, and find the max
* left/right product subarray
    1. in one for loop we essentially figure out the left product and the right product
        - left product = product from left --> right
        - right product = product from right --> left
    2. if any of our products are 0, we initialize it to the current nums[i] else we just keep the running product
    3. then we take the max of our current result, and the left/right products
* reason why we do this:
    1. Consider the case where there are no zeroes in the array
    2. If there are an even number of negatives in the array, then the max product is just the product of the entire array (because two negatives product a positive)
    3. If there are an odd number of negatives in the array, then you need to exclude one of the negative numbers* so you are left with an even number of negatives (you don't know which negative to exclude). To exclude a number, take the subarray to the left of it or to the right of it.
    4. So you need either the product of A[i] where i in [0, k - 1] or where i in [k + 1, len(A) - 1] for some k, where k in [0, len(A)] (len(a) is inclusive because you might need the entire array, e.g. if the array has length 1).
    5. You don't know which number to pivot around (i.e. the correct value of k) or which side (left or right) to take. So try all left sides (see the first loop). Then, try all right sides (see the second loop). Take the best result.
    6. Now modify this approach to work for arrays with zeroes. You just need to use the multiplicative identity 1 to "reset" after seeing a zero.
* dp algorithm:
    1. keep track of the max and min as you go along from left --> right
    2. max = Math.max(max * nums[i], min * nums[i], nums[i])
    3. min = Math.max(max * nums[i], min * nums[i], nums[i])
        - reason why we need to keep track of both max AND min is b/c of NEGATIVE numbers in the array
        - so if max = 3, min = -3, and nums[i] = -10
            * max * nums[i] = -30
            * min * nums[i] = 30
            * so the max and min have essentially switched places
        - and the reason why we also need to take a look at nums[i] for the comparison too is because the previous max/min could be 0 (due to being multiplied with a 0 previously)
            * therefore, we can just scrap that value and instead take whatever is at nums[i]

```javascript
/**
 * @param {number[]} nums
 * @return {number}
 */

 // naive algorithm
 // Time: O(n ^ 2)
 // Space: O(1)
var maxProduct = function(nums) {
    if (nums.length === 1) return nums[0];

    let max = Number.NEGATIVE_INFINITY;
    for (let i = 0; i < nums.length; i++) {
        let runningMax = nums[i];
        max = Math.max(max, runningMax);
        for (let j = i + 1; j < nums.length; j++) {
            runningMax *= nums[j];
            max = Math.max(max, runningMax);
        }
    }

    return max;
};

// Time: O(n)
// Space: O(1)
var maxProduct = function(nums) {
    if (nums.length === 1) return nums[0];

    const n = nums.length;
    let max = Number.NEGATIVE_INFINITY;

    // calculates the products from Left --> Right and Right --> Left

    /* 
     * now, why are we doing this?
     * first train of thought:
        - the largest product subarray will involve all numbers from the left or the right side unless
        one of those sides involves a 0
            * e.g. [-1, 2, 3, 1], our max subarray is [2, 3, 1] = 6
            * [-3, -2, -1, 0], our max subarray is [-3, -2]
            * [-1, 3, 2, -1], our max subarray is [-1, 3, 2, -1]
                - if we replace the nums[0] or nums[n - 1] with anything but a 0, the max product
                subarray will always involve nums[0] or nums[n - 1]
                - therefore if we do something like [-1, 3, 2, 0], then our max sub array is [3, 2]
                which doesn't involve the edges
                - therefore, if we have something like a 0 there, we can think of it like breaking up the array. instead of the edge being where 0 is at, then let the edge be after the 0
     * if we were to do a bottom-up dp algorithm, we would need this sort of loop:
        for (i: n - 1 ... 0)/
            for (j: i + 1 ... n)/
     * this essentially calculates the left and right products
        - if you look at a dp table:
            * when i = 0, leftProd = dp[0][0 ...n]
            * when j = n, rightProd = dp[n][n ... 0]
        - so the top row and the last column are the leftProd and rightProd respectively
     */
    let leftProd = 1;
    let rightProd = 1;
    
    for (let i = 0; i < n; i++) {

        // (prod === 0 ? 1 : prod) * nums[i];
        // basically, if you get a 0, then nums[i] is the product
        // else, just calculate the runningProduct
        leftProd = (leftProd === 0 ? 1 : leftProd) * nums[i];
        rightProd = (rightProd === 0 ? 1 : rightProd) * nums[n - i - 1];

        max = Math.max(max, Math.max(leftProd, rightProd));
    }

    return max;
}

// dp algorithm
// basically Kadane's algorithm with slight modification
// Time: O(n)
// Space: O(1)
var maxProduct = function(nums) {
    if (nums.length === 1) return nums[0];
    
    // in addition to keeping track of the max, we also keep track of the min
    // reason being, with the addition of negative numbers, a min * negative could yield a greater max
    // e.g. max = 3, min = -3, nums[i] = -10
    // max * nums[i] = -30
    // min * nums[i] = 30
    // therefore min > max
    let res = nums[0];
    let max = nums[0];
    let min = nums[0];

    for (let i = 1; i < nums.length; i++) {
        // the temp keeps track of the previous max since we override it
        let temp = max * nums[i];

        // the reason why we also need to compare nums[i] itself is because
        // if the previous min/max were both 0, then if nums[i] is a nonzero number
        // then it acts as both the current max and current min
        max = Math.max(temp, min * nums[i], nums[i]);
        min = Math.min(temp, min * nums[i], nums[i]);

        res = Math.max(res, max);
    }

    return res;
}
```

## Word Break

* https://leetcode.com/problems/word-break/
***
* Time Complexity:
    - naive: O(2$^{n}$ + m), where n = length of string and m = size of wordDict
        * you have 2 options: you either include the word that can break it the string or you don't = O(2$^{n}$)
        * need to create the hash table from the wordDict = O(m)
    - dp: O(n$^{2}$ + m)
        * looping from n ... 0 and checking if a word can fit into the first section of the substring = O(n$^{2}$)
        * need to create the hash table from the wordDict = O(m)
* Space Complexity:
    - naive: O(m + n)
        * need space for the hash table created from wordDict = O(m)
        * the height of the recursion tree will only be as long as the length of the string = O(n)
            - if you have a word with a single character, e.g. s = "aaaaa", and a wordDict with just one character, e.g. wordDict = ["a"], then the longest branch will be n in length
    - dp: O(m + n)
        * need space for the hash table created from wordDict = O(m)
        * need space for the dp table = O(n)
***
* naive algorithm:
    - place all words in wordDict into a hash table organized by the first letter of the word
        * wordDict = ["cat", "dog", "sand"]
        * hash = {"c": ["cat"], "d": ["dog"], "s": ["sand"]}
    - then we do a sort of dfs traversal depending on the start index
        * if the index is greater than the word, we return true b/c it is breakable
        * else we check if the current index's letter has words that start with it
            - if there aren't, then it cannot be breakable
            - but if there are, we check if the word can fit into the substring correctly and then traverse on the new start index (start + word.length)
* dp:
    - preprocess the wordDict like in the naive algorithm
    - we create a dp table of length n, where n = string's length
        * dp[i] = whether the substring[i...n] can be separated using words from the wordDict
        * initially, every index is false
    - we then loop backwards
        * if the current letter at that index does not have any words that start with it, then dp[index] = false
        * if there are, we check if the entire word can fit into the substring from [i...n]
            - if it can, then we check if the remaining portion of the substring can also be broken
            - if there is no substring remaining, meaning that the word was long enough to go from [i ... n], then it is true because it separated it cleanly
            - __recurrence relation__
                * __dp[i] = startsWith(s[i], word) and dp[i + word.length]__
                * __if dp[i + word.length] === undefined, then dp[i] = true b/c it was a clean break__
    - once we've found a word that can fit cleanly at i, we can just break from the loop
    - once the loop has ended, we return dp[0] which is the cumulation of all the word breaks from [0...n] since we looped backwards

```javascript
/**
 * @param {string} s
 * @param {string[]} wordDict
 * @return {boolean}
 */

// naive algorithm
var wordBreak = function(s, wordDict) {
    // preprocess wordDict
    const dict = {};
    wordDict.forEach(word => {
        const firstLetter = word[0];
        if (dict[firstLetter] === undefined) {
            dict[firstLetter] = [];
        }
        dict[firstLetter].push(word);
    })

    const traverse = (start) => {
        if (start >= s.length) {
            return true;
        }
        let canSeparate = false;
        let firstLetter = s[start];
        if (dict[firstLetter] === undefined) {
            return false;
        }

        for (let word of dict[firstLetter]) {
            if (s.indexOf(word, start) === start) {
                canSeparate = canSeparate || traverse(start + word.length);
            }
        }

        return canSeparate;
    }

    return traverse(0);
};

const wordMatch = (s, word, start) => {
    for (let i = 0; i < word.length; i++) {
        if (word[i] !== s[start + i]) return false;
    }

    return true;
}

// dp algorithm
// Time: O(m x n)
// Space: O(m + n)
var wordBreak = function(s, wordDict) {
    // preprocess wordDict
    const dict = {};
    wordDict.forEach(word => {
        const firstLetter = word[0];
        if (dict[firstLetter] === undefined) {
            dict[firstLetter] = [];
        }
        dict[firstLetter].push(word);
    })

    const n = s.length;

    // dp[index] represents whether the string from [index ... n] can be separated by wordDict
    const dp = Array.from({length: n}, () => false);

    for (let i = n - 1; i >= 0; i--) {
        const letter = s[i];

        // if there are no words that begin with this character
        // then it can't be separated
        if (dict[letter] === undefined) continue;

        // else we check each word that can begin with it
        for (let word of dict[letter]) {
            if (wordMatch(s, word, i)) {
                dp[i] = (i + word.length >= s.length) ? true : dp[i + word.length];
                if (dp[i]) break;
            }
        }
    }

    return dp[0];
}
```

## Longest Increasing Subsequence

* https://leetcode.com/problems/longest-increasing-subsequence/description/
***
* Time Complexity:
    - naive: O(2$^{n}$)
        * you're essentially doing a dfs/backtracking algorithm
        * you either choose to include the next largest number or you don't
    - dp: O(n$^{2}$)
        * 2 for loops traversing through the array
* Space Complexity:
    - naive: O(n)
        * needs space to create the subsequence during the traversal = O(n)
        * if the entire array is a LIS, then you'll have O(n) functions in the stack
    - dp: O(n)
        * requires space for the dp table
***
* naive algorithm:
    - it's essentially a backtracking/dfs algorithm where you either choose to include the next largest element or not
    - you can only add the current item to the subsequence if it is greater than the last element added to it, i.e. the prevMax
* dp algorithm:
    - when you have created the subsequence branches from the 0th index, the next branches for the ith...nth subsequences will have already been solved
    - so instead of doing it from 0, we can do it from the end
        * why? because we know that the last element in the array will ALWAYS have an LIS of 1.
        * you literally cannot add any more elements to the subsequence b/c there are none left. it's the end of the array
    - so we loop backwards and use this to our advantage
        * we create and initialize a dp table with all 1s
        * as we loop backwards, we have another inner loop going forwards
            - this simulates what the naive dfs algorithm was doing
            - it checks if a subsequence can be made with the current value as the head and the rest from j...n can be added to it
            - then all we have to do is compare the current LIS at i with the newly created subsequence of [i] + [j...n], where j > i
    - so our __recurrence relation__ is basically:
        * __dp[i] = max(dp[i], 1 + dp[j]) if nums[i] < nums[j] and i < j__
    - we then find the max of the entire dp table and return it
        * we do this b/c we cannot assume that the LIS starts with the first number in the array
        * this is just an O(n) check so the time complexity is still O(n$^{2}$)

```javascript
/**
 * @param {number[]} nums
 * @return {number}
 */

 // naive dfs
var lengthOfLIS = function(nums) {
    if (nums.length === 1) return 1;
    const n = nums.length;

    let max = 1;
    const traverse = (arr, index, prevMax) => {
        // basically, you create the subsequence from scratch
        // and only include the current value into the subsequence
        // if it is greater than the last number in the subsequence
        // so if we have arr = [1,2], and nums[i] = 3
        // then our arr = [1,2,3] and we traverse on it
        for (let i = index; i < n; i++) {
            if (nums[i] > prevMax) {
                arr.push(nums[i]);
                traverse(arr, i + 1, nums[i]);
                arr.pop();
            }
        }
        max = Math.max(max, arr.length);
    }

    traverse([], 0, Number.NEGATIVE_INFINITY);
    return max;
}

// dp
// Time Complexity: O(n^2)
// Space Complexity: O(n)
var lengthOfLIS = function(nums) {
    if (nums.length === 1) return 1;
    const n = nums.length;

    // dp[i] = the current LIS from i ... n
    const dp = Array.from({length: n}, () => 1);

    // we loop backwards because we know that dp[n - 1] is ALWAYS 1
    // reason being, you can't add any more values to the subsequence. the array has no more values
    for (let i = n - 1; i >= 0; i--) {
        for (let j = i + 1; j < n; j++) {
            if (nums[i] < nums[j]) {
                // since we've already checked that nums[i] < nums[j]
                // we just need to see if that gives us a bigger LIS if we add
                // nums[i] to the subsequence[j...n]
                dp[i] = Math.max(dp[i], 1 + dp[j]);
            }
        }
    }

    // we cannot assume that the LIS starts at dp[0]
    // so we have to check the entire array
    return Math.max(...dp);
};
```

## Partition Equal Subset Sum

* https://leetcode.com/problems/partition-equal-subset-sum/description/
***
* Time Complexity:
    - naive:  O(2$^{n}$)
        * basically dfs/backtracking
        * either choose to include the current item into the subset to equal total / 2 or don't include it
    - dp:  O(n * total), total = total sum of all elements in nums
        * 2 for loops looping through all elements and from 
* Space Complexity:
    - naive: O(total)
        * if the entire array is just 1s, then the function stack will have (total / 2) # of functions
    - O(total)
        * need space for the dp table
***
* naive algorithm:
    - basically employ a dfs/backtracking algorithm here
    - we count from 0 all the way up to the (total / 2)
    - if we reach that amount, then we return true
* dp algorithm:
    - we still employ the backtracking algorithm here but iteratively
    - we have 2 choices: either include the current value or don't include it to the subtotal so far
    - so our __recurrence relation__ is:
        * __dp[i] = dp[i] || dp[i - nums[i]]__
        * since we are returning a boolean value, dp[i] should also be a boolean as well
        * dp[i] represents the sum that can be made using the elements in the nums array
        * dp[i] = true if we can make that sum
    - the reason why we loop from total...0 is because we do not want to use those elements more than once like in the Coin Change algorithm
        * you can see this easily with nums = [1]
        * it would essentially make the entire dp table true

```javascript
/**
 * @param {number[]} nums
 * @return {boolean}
 */

 // naive algorithm
 // Time: O(2^n)
 // Space: O(total), total = sum of all values in nums
var canPartition = function(nums) {
    let total = 0;
    nums.forEach(val => total += val);

    // if it's an odd number, then it cannot be partitioned
    // into 2 subsets
    if (total % 2 === 1) return false;

    const traverse = (subtotal, index) => {
        if (subtotal === (total / 2)) {
            return true;
        }

        let canPart = false;
        for (let i = index + 1; i < nums.length; i++) {
            if (subtotal + nums[i] <= (total / 2)) {
                canPart = canPart || traverse(subtotal + nums[i], i);
            }
        }

        return canPart;
    }

    return traverse(0, 0, nums[0]);
};

// dp
// Time: O(n^2)
// Space: O(n)
var canPartition = function(nums) {
    let total = 0;
    nums.forEach(val => total += val);

    // if it's an odd number, then it cannot be partitioned
    // into 2 subsets
    if (total % 2 === 1) return false;

    total /= 2;

    // dp[i] represents if we can make that amount, i, with the given elements in nums
    // dp[0] is always true b/c we can make 0 with 0 elements
    const dp = Array.from({length: total + 1}, () => false);
    dp[0] = true;

    for (let num of nums) {
        for (let sum = total; sum >= 0; sum--) {
            if (num <= sum) {
                // we can either already create the sum
                // or we need to use the current num to get to the sum
                // so if num = 7 and sum = 11, we need to see if we were able to make 4 before
                
                // this is essentially the 0/1 knapsack
                // where we either include the current num or not include it
                // dp[sum] = do not include
                // dp[sum - num] = include

                // since we need to return a boolean,
                // our recurrence relation does not rely on maxes/mins
                dp[sum] = dp[sum] || dp[sum - num];
            }
        }
    }

    return dp[total];
}
```

## Triangle

* https://leetcode.com/problems/triangle/description/
***
* Time Complexity:
    - naive: O(2$^{n}$)
        * for every node, you have two options: [r + 1][c] or [r + 1][c + 1]
        * since this is a triangle, n = numRows === numCols, and we only recursively traverse the triangle until we reach the bottom
    - topdown memo: O(n$^{2}$)
        * by implementing memoization, we reduce the number of subproblems to O(n$^{2}$)
    - bottom-up dp: O(n$^{2}$)
        * we use a nested for loop that traverses through every value in the triangle
* Space Complexity:
    - naive: O(n)
        * uses recursion so needs space for the function calls in the stack
        * the depth of the recursion is bounded by the length of the rows since we go from the top of the triangle to the bottom
        * we return once we reach the bottom of the triangle so it'll only have O(n) function calls
    - topdown memo: O(n$^{2}$)
        * requires space for the dp table which has (r * c) or n$^{2}$ cells
    - bottom-up dp: O(n$^{2}$)
        * requires space for the dp table which has (r * c) or n$^{2}$ cells
    - bottom-up dp - space optimized: O(n)
        * we only care about dp[r + 1][c] and dp[r + 1][c + 1]
        * the only dimension that changes is c since we always keep [r + 1]
        * thus, we can remove that dimension entirely and make it O(n)
            - in a triangle, num rows === num cols
***
* naive:
    - basically just dfs traversal from top to bottom
* topdown memo:
    - since the algorithm can go [r + 1][c] and [r + 1][c + 1], there are some instances where we'd see overlap
    - so instead of recalculating that sum, we just memoize and return it
* bottom-up dp:
    - from the naive algorithm, we know that our __recurrence relation__ is:
        * __dp[r][c] = triangle[r][c] + Math.min(dp[r + 1][c], dp[r + 1][c + 1])__
    - using this recurrence relation, we can also determine how we want to structure our nested for loops
        * since the algorithm relies on [r + 1] we know our outer loop would need to go from [n - 2...0]
        * reason why we go from n - 2 is b/c we have already initialized the last row of our dp table with the last row from triangle since that's the default values we are starting from
        * as for c, we rely on [c] and [c + 1] which means we also have to work backwards as well so we go from [c...0]
            - but since this is a triangle shape, the column length will be determined by what row we're in
            - e.g. row 2 = 2 cols
    - once we've finished filling up the dp table, we should return dp[0][0] since we're working our way up from the bottom of the triangle
* bottom-up dp - space optimized:
    - notice how, in our recurrence relation, the only dimension that doesn't change is [r + 1]
    - since we only need later rows, we can drop this dimension from our recurrence relation
    - so the __space optimized recurrence relation is:__
        * __dp[c] = triangle[r][c] + Math.min(dp[c], dp[c + 1])__
    - since our recurrence relation removed the extra row dimension, we must also change the way we loop
        * the outer loop is fine since we implicitly go from bottom-top, our outer loop must do the same
        * thus the inner loop handling the column must be changed
            - this is b/c it relies on [c] AND [c + 1]
            - if we loop from [c...0], we actually override the implicit values from the [r + 1] row
            - so we have to loop from [0...c] to make sure that [c + 1] is the value from the [r + 1] row

```javascript
// correct naive
var minimumTotal = function(triangle) {
    const n = triangle.length;
    const traverse = (r, c) => {
        if (r === n) {
            return 0;
        }

        let left = triangle[r][c] + traverse(r + 1, c);
        let right = triangle[r][c] + traverse(r + 1, c + 1);

        return Math.min(left, right);
    }

    return traverse(0, 0);
}

// correct topdown memo
var minimumTotal = function(triangle) {
    const n = triangle.length;
    const seen = {};
    const traverse = (r, c) => {
        if (r === n) {
            return 0;
        }

        const key = `${r}_${c}`;
        if (seen[key] !== undefined) {
            return seen[key];
        }

        let left = triangle[r][c] + traverse(r + 1, c);
        let right = triangle[r][c] + traverse(r + 1, c + 1);


        seen[key] = Math.min(left, right);
        return seen[key];
    }

    return traverse(0, 0);
}

// bottom-up dp
var minimumTotal = function(triangle) {
    const numRows = triangle.length;
    const dp = Array.from({ length: numRows - 1 }, () => []);
    dp[numRows - 1] = [...triangle[numRows - 1]];

    for (let r = numRows - 2; r >= 0; r--) {
        for (let c = triangle[r].length - 1; c >= 0; c--) {
            dp[r][c] = triangle[r][c] + Math.min(dp[r + 1][c], dp[r + 1][c + 1]);
        }
    }

    return dp[0][0];
}

// bottom-up dp space optimized
// for a triangle, numRows == numCols
var minimumTotal = function(triangle) {
    const numRows = triangle.length;
    const dp = [...triangle[numRows - 1]];

    for (let r = numRows - 2; r >= 0; r--) {
        for (let c = 0; c < triangle[r].length; c++) {
            dp[c] = triangle[r][c] + Math.min(dp[c], dp[c + 1])
        }
    }
    return dp[0];
}
```
